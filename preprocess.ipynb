```python
import json
```

```python
file = "outputs/Etiketlendi-trendyol_turkcell_superbox_yorumlar_159_adet.json"
```

```python
def read_json(path:str):
    with open(path,"r",encoding="utf-8") as f:
        data = json.load(f)
    return data
```

```python
json_data = read_json(file)
```

```python
for i, item in enumerate(json_data):
    entities = item["entities"]
    for j in range(len(entities) - 1, -1, -1):
        entity = entities[j]
        if entity["label"].lower() in ["olumlu", "olumsuz", "nötr"]:
            json_data[i]["entities"].pop(j)
```

```python
def process(filepath:str) -> None:
    json_data = read_json(filepath)
    for i, item in enumerate(json_data):
        entities = item["entities"]
        for j in range(len(entities) - 1, -1, -1):
            entity = entities[j]
            if entity["label"].lower() in ["olumlu", "olumsuz", "nötr"]:
                json_data[i]["entities"].pop(j)
    with open(filepath.replace("original/","data/"),"w",encoding="utf-8") as f:
        json.dump(json_data,f,ensure_ascii=False,indent=4)
```

```python
process("original/trendyol_superbox.json")
process("original/turkcell_sikayetvar.json")
process("original/vodafone.json")
process("original/yorum_budur_turkcell.json")
```
